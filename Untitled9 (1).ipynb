```python
import statsmodels.api as sm
x1 = sm.add_constant(x)
results = sm.OLS(y,x1).fit()
results.summary()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,Y_test = train_test_split(x,y, test_size= 0.3, train_size = 0.7)
```

```python
Model = []
Accuracy = []
F1Score = []
```

```python
from sklearn.linear_model import LogisticRegression

log_re = LogisticRegression(class_weight='balanced')
log_re.fit(X_train, y_train)


y_pred = log_re.predict(X_test) 
ypred_prob_logreg = log_re.predict_proba(X_test)


yhat_positive_logreg = ypred_prob_logreg[:, 1]
```

```python
from sklearn.metrics import classification_report,f1_score
print(classification_report(Y_test, y_pred))
print()
print (pd. crosstab(Y_test, y_pred))
print ()
fpr, tpr, thresholds = roc_curve(Y_test, yhat_positive_logreg)
auc = roc_auc_score(Y_test, yhat_positive_logreg)
print('AUC: %.3f' % auc)
accuracy=log_re.score(X_test,Y_test)
print("accuracy","=",accuracy*100,"%")
print ()
scores = cross_val_score(log_re, x, y, cv=10)
scores
print ("cross validation score","=",scores.mean()*100,"%")
Model.append('LogisticRegression')
F1Score.append(f1_score(Y_test, y_pred, average=None))
Accuracy.append(accuracy)



print("Confusion Matrix for Logistic Regression Classifier","\n")
score = round(accuracy_score(Y_test, y_pred),3) 
cm1 = confusion_matrix(Y_test, y_pred)
sns.heatmap(cm1, annot=True, fmt=".1f", linewidths=.3, 
        square = True, cmap = 'Greens')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.title('Accuracy Score: {0}'.format(score), size = 12)
plt.show()
print("\n")
```

```python
from sklearn.tree import DecisionTreeClassifier
dec_tr = DecisionTreeClassifier(class_weight='balanced')
dec_tr.fit(X_train, y_train)

y_pred = dec_tr.predict(X_test) 
ypred_prob_dt = dec_tr.predict_proba(X_test) 

yhat_positive_dt = ypred_prob_dt[:, 1]
```

```python
from sklearn.metrics import classification_report,f1_score
print(classification_report(Y_test, y_pred))
print()
print (pd. crosstab(Y_test, y_pred))
print ()
fpr1, tpr1, thresholds1 = roc_curve(Y_test, yhat_positive_dt)
auc1 = roc_auc_score(Y_test, yhat_positive_dt)
print('AUC: %.3f' % auc)
accuracy=dec_tr.score(X_test,Y_test)
print("accuracy","=",accuracy*100,"%")
print ()
scores = cross_val_score(dec_tr, x, y, cv=10)
scores
print ("cross validation score","=",scores.mean()*100,"%")
Model.append('DecisionTreeClassifier')
F1Score.append(f1_score(Y_test, y_pred, average=None))
Accuracy.append(accuracy)


print("Confusion Matrix for Decision Tree Classifier","\n")
score = round(accuracy_score(Y_test, y_pred),3) 
cm1 = confusion_matrix(Y_test, y_pred)
sns.heatmap(cm1, annot=True, fmt=".1f", linewidths=.3, 
        square = True, cmap = 'Greens')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.title('Accuracy Score: {0}'.format(score), size = 12)
plt.show()
print("\n")
```
